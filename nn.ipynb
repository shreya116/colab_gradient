```
# Training loop 
for epoch in range(100):
  for x_val, y_val in zip(x_data, y_data): 
    y_pred = model(x_val)
    l = criterion(y_pred, y_val)
    print(l)
    print(epoch, l.data)
    optimizer.zero_grad()
    l.backward(retain_graph = True)
    optimizer.step()
    #print(w1.grad)
    #print("\tgrad: ", x_val, y_val, w1.grad.item,w2.grad.item,b.grad.item)

  print("progress:", epoch, "weight:",{list(model.parameters())[0].data}, "loss=", l.data) 
```
